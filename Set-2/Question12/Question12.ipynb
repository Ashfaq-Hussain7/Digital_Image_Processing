```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load Image (Grayscale)

img = cv2.imread("image.jpg", cv2.IMREAD_GRAYSCALE)
img = cv2.resize(img, (256, 256))
```

```python
# Define Line Detection Masks

# Horizontal mask
horizontal_mask = np.array([[-1, -1, -1],
                            [ 2,  2,  2],
                            [-1, -1, -1]])

# Vertical mask
vertical_mask = np.array([[-1,  2, -1],
                          [-1,  2, -1],
                          [-1,  2, -1]])

# Diagonal 45° mask
diag1_mask = np.array([[ 2, -1, -1],
                       [-1,  2, -1],
                       [-1, -1,  2]])

# Diagonal 135° mask
diag2_mask = np.array([[-1, -1,  2],
                       [-1,  2, -1],
                       [ 2, -1, -1]])
```

```python
# Apply Filters

horizontal_lines = cv2.filter2D(img, -1, horizontal_mask)
vertical_lines   = cv2.filter2D(img, -1, vertical_mask)
diag1_lines      = cv2.filter2D(img, -1, diag1_mask)
diag2_lines      = cv2.filter2D(img, -1, diag2_mask)
```

```python
# Display Results

titles = ["Original", "Horizontal Lines", "Vertical Lines", "Diagonal 45°", "Diagonal 135°"]
images = [img, horizontal_lines, vertical_lines, diag1_lines, diag2_lines]

plt.figure(figsize=(12,6))
for i in range(5):
    plt.subplot(2,3,i+1)
    plt.imshow(images[i], cmap="gray")
    plt.title(titles[i])
    plt.axis("off")

plt.tight_layout()
plt.show()
```
