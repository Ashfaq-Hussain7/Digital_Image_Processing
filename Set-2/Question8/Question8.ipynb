```python
# sharpen_rgb_laplacian.ipynb

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Function to display images
def show_images(images, titles):
    plt.figure(figsize=(12, 6))
    for i, (img, title) in enumerate(zip(images, titles)):
        plt.subplot(1, len(images), i + 1)
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        plt.title(title)
        plt.axis("off")
    plt.tight_layout()
    plt.show()
```

```python
# Load image
img = cv2.imread("image.jpg")   # replace with your image filename
if img is None:
    raise FileNotFoundError("Image not found! Please put 'sample.jpg' in the working directory.")

show_images([img], ["Original RGB Image"])
```

```python
# Apply Laplacian filter
laplacian = cv2.Laplacian(img, cv2.CV_64F)   # Laplacian detects edges
laplacian = cv2.convertScaleAbs(laplacian)   # convert back to 8-bit
```

```python
# Sharpen the image (Original - Laplacian)
sharpened = cv2.addWeighted(img, 1.0, laplacian, -1.0, 0)
```

```python
# Display results
show_images([img, laplacian, sharpened],
            ["Original", "Laplacian (Edges)", "Sharpened with Laplacian"])
```
