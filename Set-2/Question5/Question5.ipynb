```python
# Import Libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.util import random_noise
```

```python
# Load the image
image = cv2.imread("sample.jpg", cv2.IMREAD_GRAYSCALE)

if image is None:
    raise FileNotFoundError("Image not found! Please check the path.")

plt.figure(figsize=(5,5))
plt.title("Original Image")
plt.imshow(image, cmap="gray")
plt.axis("off")
plt.show()
```

```python
# Add noise
noisy_img = random_noise(image, mode='s&p', amount=0.05)  # 5% noise
noisy_img = (noisy_img * 255).astype(np.uint8)  # Convert back to uint8

plt.figure(figsize=(5,5))
plt.title("Image with Salt and Pepper Noise")
plt.imshow(noisy_img, cmap="gray")
plt.axis("off")
plt.show()
```

```python
# Apply Median Filtering
median_filtered = cv2.medianBlur(noisy_img, 3)  # Kernel size 3x3

plt.figure(figsize=(5,5))
plt.title("Median Filtered Image")
plt.imshow(median_filtered, cmap="gray")
plt.axis("off")
plt.show()

# Cell 5: Compare Results
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
titles = ["Original", "Noisy (Salt & Pepper)", "Median Filtered"]
images = [image, noisy_img, median_filtered]

for i in range(3):
    axs[i].imshow(images[i], cmap='gray')
    axs[i].set_title(titles[i])
    axs[i].axis('off')

plt.show()
```
