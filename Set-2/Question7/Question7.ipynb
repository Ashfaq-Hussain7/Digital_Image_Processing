```python
# smoothing_rgb.ipynb

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Function to display images
def show_images(images, titles):
    plt.figure(figsize=(12, 6))
    for i, (img, title) in enumerate(zip(images, titles)):
        plt.subplot(1, len(images), i + 1)
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        plt.title(title)
        plt.axis("off")
    plt.tight_layout()
    plt.show()
```

```python
# Load an RGB image
img = cv2.imread("sample.jpg")   # change filename if needed
if img is None:
    raise FileNotFoundError("Image not found! Please put 'sample.jpg' in the working directory.")

show_images([img], ["Original Image"])
```

```python
# Apply linear spatial filters (smoothing)
# Mean filter (Averaging filter)
smoothed_3x3 = cv2.blur(img, (3, 3))   # 3x3 kernel
smoothed_5x5 = cv2.blur(img, (5, 5))   # 5x5 kernel
smoothed_11x11 = cv2.blur(img, (11, 11))  # 11x11 kernel
```

```python
# Display results
show_images(
    [img, smoothed_3x3, smoothed_5x5, smoothed_11x11],
    ["Original", "Smoothed (3x3)", "Smoothed (5x5)", "Smoothed (11x11)"]
)
```
